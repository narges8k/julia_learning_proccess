```julia
using DifferentialEquations
using Plots
S0=0.999
I0=0.001
INPUT=[S0; I0; 0]
gamma=1/7
betta=1.428
tend=90
```

```julia
function equations(du, u, p, t)
    du[1]=-betta*u[1]*u[2]
    du[2]=betta*u[1]*u[2]-gamma*u[2]
    du[3]=gamma*u[2]
    end
```

```julia
tspan=[0.0,tend]
problem=ODEProblem(equations, INPUT, tspan)
solution=solve(problem)
```

```julia
pS=plot(solution, vars=(0,1), label="S")
pI=plot(solution, vars=(0,2), label="I")
pR=plot(solution, vars=(0,3), label="R")
plot(pS,pI,pR)
```
